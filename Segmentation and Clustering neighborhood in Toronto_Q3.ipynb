```python
#Import Beautiful Soup to scrap the web
from bs4 import BeautifulSoup
import requests
```

```python
#Get from the website
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup (source, 'html.parser')
#Find the table of post code in the website and read into pandas table
post_code_table = soup.find('table', class_='wikitable sortable')
df = pd.read_html(str(post_code_table), header = 0)[0]
df.rename(columns = {'Postcode' : 'PostalCode'}, inplace = True)
df
```

```python
# Drop rows which Borough is not assigned.
df.drop(df[df.Borough == 'Not assigned'].index, inplace=True)
df
```

```python
# Combine Neighbourhood into 1 cell if the postcode is the same
df = df.groupby(['PostalCode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df
```

```python
# Find the Borough where the Neighbourhood is not assigned
temp= df.loc[df.Neighbourhood == 'Not assigned']
temp.Borough
```

```python
#Replace with the value of the Borough for the unassigned Neighbourhood
df.Neighbourhood.replace('Not assigned',temp.Borough, inplace = True)
df
```

# Part 3

```python
#Import libraries for data visualization
import geopy
import folium
```

```python
#
toronto_data = geo_df[geo_df['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data.head()
```

```python
#Get geographical coordinate of Toronto
address = 'Toronto'
geolocator = Nominatim(user_agent="trt_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
#This is the first neighbourhood's name
toronto_data.loc[0,'Neighbourhood']
```

```python
# set number of clusters
kclusters = 5

toronto_clustering = toronto_data.drop(['Neighbourhood', 'PostalCode','Borough'],1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:50] 
```

```python
# add clustering labels into the toronto_data
toronto_data.insert(0,'Cluster Labels',kmeans.labels_)
toronto_data
```

```python
#Visualize the resulting clusters
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood'], toronto_data['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster],
        fill=True,
        fill_color=rainbow[cluster],
        fill_opacity=0.9).add_to(map_clusters)
       
map_clusters
```
